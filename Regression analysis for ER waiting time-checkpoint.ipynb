# Regression Analysis
This part of our case study on ER time optimisation will be on measuring the affecting
factors on the waitng time and patient outcome. Starting by importing libararies and loading our dataset


## Objectives:
1. Identify correlations between factors
2. Biuld an OLS model to analys the effect of diffrent variables on wait time
3. Build a predictive model to predict future wait times
4. Build a RFclassifer to predict future ratings of seviece  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, f1_score,  recall_score, r2_score, mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv("cleanERdata.csv", sep =',')
```

```python

df['hour'] = df['time'].str[:1].astype(np.int64)
print(df.columns,df.shape,df.dtypes)
```

## Addressing correlations between wait time and numerical and categorical features

```python
#numerical correlations 

numericals = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numericals].corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix of Numerical Factors')
plt. tight_layout()
plt.show()
```

```python
# calculating correlation
factors = pd.get_dummies(df.drop(columns=['patient_satisfaction','visit_date','hospital_name','patient_id','visit_id','data_flag']),drop_first =True)
correlations = factors.corr()['total_wait_time']

print(correlations)
```

```python
factors = np.asarray(pd.get_dummies(df[['patient_number','nurse_to_patient_ratio','urgency_level', 'month','time_of_day']],columns =['urgency_level', 'month', 'time_of_day'],drop_first = True)).astype(np.int64)
Y = np.log1p(df['total_wait_time']).values
factors = sm.add_constant(factors)
#X_train, X_test, Y_train,Y_test = train_test_split(factors, Y, test_size=0.2)

model = sm.OLS(Y, factors)
result = model.fit()
result.summary()
```

```python
residuals = result.resid
fitted = result.fittedvalues


fig, axes = plt.subplots(2,2,figsize=(12,10))

axes[0,0].scatter(fitted,residuals, alpha =0.7)
axes[0,0].axhline(y=0,color='red', alpha =0.7)
axes[0,0].set_xlabel('Fitted Values')
axes[0,0].set_ylabel('Residuals')
axes[0,0].set_title('Residuals VS Fitted values')


sm.qqplot(residuals, line ='45',ax = axes[0,1], fit = True)
axes[0,1].set_title('Q-Q plot')


axes[1,0].hist(residuals, bins = 20 , alpha=0.7)
axes[1,0].set_xlabel('Residuals')
axes[1,0].set_ylabel('Frequency')
axes[1,0].set_title('Distribution of Residuals')


axes[1,1].scatter(fitted, np.sqrt(np.abs(residuals)))
axes[1,1].set_xlabel('Fitted Values')
axes[1,1].set_ylabel('Standarised Residuals')
axes[1,1].set_title('Scale-Location Plot')





plt.tight_layout()
plt.show()
```

## Linear model to predict future wait times

```python
from sklearn.linear_model import LinearRegression


X_train, X_test, Y_train,Y_test = train_test_split(factors, Y, test_size=0.3)



model = LinearRegression()
results = model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
intercept = model.intercept_
ceof = model.coef_

print('===R2-score===',r2_score(Y_test,Y_pred))

print('===MAE===',mean_absolute_error(Y_test,Y_pred))

print('===MSE===',mean_squared_error(Y_test,Y_pred))

print('===Intercept===',intercept)

print('===coefficients====',ceof)


Y_test = np.asarray(Y_test)
Y_pred = np.asarray(Y_pred)
result = [Y_pred,Y_test]
x = len(Y_test)


print('\n')

for i in range(x):
    
    
    result = {
        'original value': Y_test,
        'predicted value': Y_pred,
        'R2-score': r2_score(Y_test,Y_pred),
        'MSE': mean_squared_error(Y_test,Y_pred),
        'MAE': mean_absolute_error(Y_test,Y_pred)
    }
m = pd.DataFrame(result)
wait_results = m.to_csv('Wait Time LR Model.csv') 


print(m.head(4))
```

## Patient satisfaction prediction model 

```python
#One hot encoding and getting right shape  
factors = pd.get_dummies(df.drop(columns=['visit_date','time','hospital_name','patient_id','visit_id','data_flag']),drop_first =True)
factors = factors.astype(np.int64)
factors.shape
```

```python
X = factors
Y = df['patient_satisfaction']
```

```python
# Spliting data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)


# Model fitting
rate_model = RandomForestClassifier(random_state=42)
results = rate_model.fit(X_train,Y_train)

# Prediction
Y_pred = rate_model.predict(X_test)

# Model evaluation
print(' == accuracy == ', accuracy_score(Y_test,Y_pred))
print('== classification report ==\n ', classification_report(Y_test,Y_pred))
print('== confusion matrix ==\n ',confusion_matrix(Y_test, Y_pred))
print(' == R2 score == ', r2_score(Y_test,Y_pred))
print('== MAE ==\n ', mean_absolute_error(Y_test,Y_pred))
print('== RSME ==\n ',np.sqrt(mean_squared_error(Y_test, Y_pred)))
```

```python
cm = confusion_matrix(Y_test, Y_pred)

n_classes = cm.shape[0]
metrics = {}


for i in range(n_classes):
    TP = cm[i,i]
    FP = cm[:,i].sum() - TP
    FN = cm[i,:].sum() - TP
    TN = cm.sum() - (TP + FP + FN)
    
    metrics.append({
        'class': [i for i in range(1,6)][i] ,
        'True positives': TP,
        'True negatives': TN,
        'False positives': FP,
        'False negatives': FN
    })

performance = pd.DataFrame(metrics)
performance['Percision'] = precision_score(Y_test, Y_pred, average = None)
performance['F1 score'] = f1_score(Y_test,Y_pred,average = None)
performance['Recall'] = recall_score(Y_test, Y_pred, average = None)
model_results = performance.to_csv('Ratings_ModelResult.csv')
```
